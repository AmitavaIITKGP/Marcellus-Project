# Individual Decision Trees

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from IPython.display import display
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel("Model_Data.xlsx")
```

```python
data.describe()
```

```python
data['Date'] = pd.to_datetime(data['Date'], format = '%Y-%m-%d')
```

```python
data['Date'].dtype
```

```python
index = data[(data['Date']>='2020-02-01') & (data['Date']<='2020-06-30')].index
```

```python
data_excluded = data.drop(index)
```

```python
data_excluded['Date'].dtype
```

```python
data_excluded.columns[1:8]
```

```python
data_excluded.describe()
```

```python
def individual_trees_excluded(testsize=0.2): 
    for score in data_excluded.columns[1:8]:
        X = data_excluded[score]
        y = data_excluded['0_Day_Lag'].apply(lambda x : 1 if x>0.01 else 0)
        X_train,X_test,y_train,y_test = X.iloc[int(testsize*len(X)):], X.iloc[:int(testsize*len(X))], y.iloc[int(testsize*len(y)):], y.iloc[:int(testsize*len(y))]
        X_train = np.array(X_train).reshape(-1,1)
        y_train = np.array(y_train).reshape(-1,1)
        X_test = np.array(X_test).reshape(-1,1)
        y_test = np.array(y_test).reshape(-1,1)
        treeClassifier = DecisionTreeClassifier(max_depth=3)
        treeClassifier.fit(X_train,y_train)
        y_pred = treeClassifier.predict(X_test)
        report = classification_report(y_test,y_pred)
        print("Decision Tree Report Based on "+score+" :")
        print(report)
        data_excluded['Prediction_'+score]=pd.Series(dtype=int)
        data_excluded['Prediction_'+score].iloc[:len(y_pred)] = y_pred
        print("Model Predictions Based on "+score+" :")
        display(data_excluded.groupby('Prediction_'+score)['0_Day_Lag'].describe()) #Based on Model Predictions
    
    return
```

```python
def customized_tree(features,testsize=0.2):
    df = data.drop(index)
    X = df[features]
    y = df['0_Day_Lag'].apply(lambda x : 1 if x>0.01 else 0)
    X_train,X_test,y_train,y_test = X.iloc[int(testsize*len(X)):], X.iloc[:int(testsize*len(X))], y.iloc[int(testsize*len(y)):], y.iloc[:int(testsize*len(y))]
    y_train = np.array(y_train).reshape(-1,1)
    y_test = np.array(y_test).reshape(-1,1)
    treeClassifier = DecisionTreeClassifier(max_depth=3)
    treeClassifier.fit(X_train,y_train)
    y_pred = treeClassifier.predict(X_test)
    report = classification_report(y_test,y_pred)
    print("Decision Tree Report Based on Customized Features :")
    print(report)
    df['Predictions']=pd.Series(dtype=int)
    df['Predictions'].iloc[:len(y_pred)] = y_pred
    print("Model Predictions Based on Customized Features :")
    display(df.groupby('Predictions')['0_Day_Lag'].describe()) #Based on Model Predictions
    
    return
```

# RANDOM FOREST CLASSIFIER

```python
# import important packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import classification_report
#import pandas_profiling

from matplotlib import rcParams
import warnings

warnings.filterwarnings("ignore")

# figure size in inches
rcParams["figure.figsize"] = 10, 6
np.random.seed(42)
```

```python
feature_data = pd.read_excel("Model_Data.xlsx")
```

```python
feature_data['Date'] = pd.to_datetime(feature_data['Date'], format = '%Y-%m-%d')
```

```python
index = feature_data[(feature_data['Date']>='2020-02-01') & (feature_data['Date']<='2020-06-30')].index
```

```python
feature_data = feature_data.drop(index)
```

```python
feature_data['Signal'] = feature_data['0_Day_Lag'].apply(lambda x:1 if x>0.01 else 0)
```

```python
feature_data.columns
```

```python
X = feature_data.iloc[22:,1:8]
y = feature_data.iloc[22:,-1]
```

```python
X_train,X_test,y_train,y_test = X.iloc[int(0.2*len(X)):], X.iloc[:int(0.2*len(X))], y.iloc[int(0.2*len(y)):], y.iloc[:int(0.2*len(y))]
```

```python
# create the classifier
classifier = RandomForestClassifier(n_estimators=100)

# Train the model using the training sets
classifier.fit(X_train, y_train)
```

```python
# predictin on the test set
y_pred = classifier.predict(X_test)
```

```python
# Calculate Model Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
report = classification_report(y_test,y_pred)
print(report)
```
